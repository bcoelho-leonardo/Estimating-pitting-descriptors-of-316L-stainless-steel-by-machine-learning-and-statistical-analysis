```python
def coefficient_of_variation(data):
    CV =[]
    mean=np.mean(data)
    std=np.std(data)
    CV=round(100*std/mean,1)
    return CV
```

### entire data

#### Epit

```python
CV1=coefficient_of_variation(pit_pass_log_0005_100["Epit_x"])
CV2=coefficient_of_variation(pit_pass_log_001_100["Epit_x"])
CV3=coefficient_of_variation(pit_pass_log_001_50["Epit_x"])
CV4=coefficient_of_variation(pit_pass_log_005_100["Epit_x"])
CV5=coefficient_of_variation(pit_pass_log_005_50["Epit_x"])
[i for i in [CV1,CV2,CV3,CV4,CV5]] 
```

### top 4 highest log(j) Vs E curves superimposed with their Epit/log(jpit) estimates

```python
# Function calculates the mean residuals of the log(j) values, selecting the top 4 highest curves
def top_highest_residuals(j, num_top_curves=6):
    mean = j.mean(axis=1)
    res = j.subtract(mean, axis=0)
    res = res.abs()
    sum_res = res.sum(axis=0)
    sum_res_sorted = sum_res.sort_values(ascending=False)  # Sort in descending order
    top_curves_indices = sum_res_sorted.index[:num_top_curves]
    return j[top_curves_indices]

# Usage:
num_top_curves = 6  # You can change this to select a different number of top curves
sum_res_0005_100 = top_highest_residuals(df_0005_100_log, num_top_curves)
sum_res_001_100 = top_highest_residuals(df_001_100_log, num_top_curves)
sum_res_001_50 = top_highest_residuals(df_001_50_log, num_top_curves)
sum_res_005_100 = top_highest_residuals(df_005_100_log, num_top_curves)
sum_res_005_50 = top_highest_residuals(df_005_50_log, num_top_curves)
```
